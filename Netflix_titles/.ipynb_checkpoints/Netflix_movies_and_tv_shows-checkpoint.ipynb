해당 데이터는 2019년에 넷플릭스에서 이용가능했던 tv 프로그램과 영화의 데이터 셋이다. 총 12개 칼럼과 7787 행으로 구성되어 있다.

해당 데이터의 컬럼을 자세히 살펴보면 다음과 같다.
- show_id: 모든 tv 프로그램과 영화의 unique ID
- type: Movie / TV Show 타입 분류
- title: 프로그램 이름
- director: 영화 감독 이름
- cast: 영화 / tv 프로그램에 출연한 인물
- country: 제작 국가
- date_added: 영화 / tv 프로그램이 넷플릭스에 추가된 날짜
- release_year: 영화 / tv 프로그램이 실제 상영(방영)된 날짜
- rating: TV 평가
- duration: 전체 지속 기간(분 / 시즌)
- listed_in: 장르
- description: 요약 설명
