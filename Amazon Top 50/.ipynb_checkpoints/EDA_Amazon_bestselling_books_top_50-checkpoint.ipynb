```python
from pandas import Series, DataFrame

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../Data_source/bestsellers with categories.csv')
df.head()
```

```python
df = DataFrame(df, columns=['Year', 'Name', 'Author', 'User Rating', 'Reviews', 'Genre', 'Price'])

df['Year'] = df['Year'].astype('category')
df.dtypes
```

```python
df.info()
```

```python
# df index name 지칭

df.index.name = 'No.'
df.columns.name = 'Info.'
df.head()
```

```python
# 년도별 내림차순

df = df.sort_values(by='Year', ascending=False)
df.head()
```

```python
# 전체 장르(fiction, non fiction) 비율

grouped_genre = df.groupby(df['Genre']).size()
print(grouped_genre)
print('===============')

ratio_genre = round(df['Genre'].value_counts(normalize=True),2)
ratio_genre
```

11년간 베스트셀러 50의 책, 550권 중 논픽션 책이 약 56%로 픽션 책보다 더 많은 비중을 차지.

```python
# 년도별 장르 비율 파악

grouped_genre_year = df.groupby(['Year', 'Genre']).size()
print(grouped_genre_year)

sns.set(rc={'figure.figsize':(9.7,7.27)})
sns.countplot(x=df['Year'], hue=df['Genre'], palette='Set2')
plt.show()
```

1. 전반적으로는 논픽션이 픽션보다 더 많은 비중을 차지.
2. 2009-2019 연도별로 보면 2015년이 두 장르 간 베스트 셀러 책의 차이가 가장 큼(17-33).
3. 픽션 책은 2014년에 29권으로 가장 많이 팔렸고, 논픽션 책은 2015년에 33권으로 가장 많이 팔림. 반면 2015년은 픽션 책이 17권으로 가장 적게 팔리고, 2014년은 논픽션 책이 21권으로 가장 적게 팔림
4. 대부분 매년 논픽션 책이 픽션 책보다 더 많은 비중을 차지. 2015년만 제외.

- 2015년엔 어떤 논픽션 책들이 있었을까? 리뷰나 평점이 높은 책들은 무엇이었을까? 
- 2014년엔 어떤 픽션 책들이 있었을까? 리뷰나 평점이 높은 책들은 무엇이었을까?

```python
# 2015년엔 어떤 논픽션 책들이 있었나

np_2015 = df[(df.iloc[:, 0] == 2015) & (df.iloc[:, 5] == 'Non Fiction')]
np_2015.shape     # 33권
np_2015.sort_values(by='User Rating', ascending=False)
```

```python
# 평점 분포 그래프

sns.countplot(x=np_2015['User Rating'], palette='Set2').set_title('the distribution graph of non fiction user rating in 2015')
plt.show()
```

    1. 33권의 책 모두 4.0 이상의 평점으로 'StrengthsFinder2.0'을 제외하면 4.4 이상이며, 대부분 4.6~4.8 평점에 분포해 있다.
    2. 약 8권의 책이 '컬리링 북'으로 2015년도에는 컬러링 북이 인기가 있었음을 추측해볼 수 있다. 평점도 4.4 이상으로 높은 평점을 받았다.
    3. 그 외에 요리책, 심리책과 같은 도서도 눈에 띈다.

```python
# 2014년도에는 어떤 픽션 책들이 있었나

p_2014 = df[(df.iloc[:, 0] == 2014) & (df.iloc[:, 5] == 'Fiction')]
p_2014.shape     # 29권

print("the popular authors of fiction in 2014:\n", p_2014.value_counts(p_2014['Author'])[:3])
p_2014.sort_values(by='User Rating', ascending=False)
```

```python
# 평점 분포 그래프

sns.countplot(x=p_2014['User Rating'], palette='Set2').set_title('the distribution graph of fiction user rating in 2014')
plt.show()
```

  1. 'John Green'의 책 3권이 2014년 베스트셀러 50에 들어갔다. 이 중 'the fault in our stars' 책 2권이 올라갔는데 가격이 다른 것으로 보아 다른 책인 것으로 보인다. 그리고 'Veronica Roth', 'RH Disney'의 책이 각각 2권씩 50 안에 들어갔다.  
  2. 평점이 3.9 - 4.9로 다양하게 분포해 있으며, 4.6 - 4.8에 다수의 책이 포함되어 있다. 

```python
# 베스트셀러인 책의 저자들 중 가장 많은 책을 보유한 저자는 누구일까? ==> 그 저자의 책은 현재에도 베스트셀러일까? 시간이 흘러도 사랑받는?

authors = df.value_counts(df['Author'])
authors[:5]
```

```python
# 'Publication Manual of the American Psychological Association, 6th Edition' 책은 한 해를 빼고 매년 베스트셀러 50에 들어감

psychological = df[df.iloc[:, 1] == 'Publication Manual of the American Psychological Association, 6th Edition']
psychological
```

글을 집필함에 있어 필요한 매뉴얼에 대한 책으로 10년간 베스트셀러 50에 들어간 것을 보면 미국인들이 글을 쓰는 매뉴얼 등에 대해 관심이 크다는 것을 추측해볼 수 있음

```python
# 'StrengthsFinder 2.0' 책은 9번 베스트셀러 50에 들어감

strengths = df[df.iloc[:, 1] == "StrengthsFinder 2.0"]
strengths
```

 1. 평점이 4.0으로 다른 베스트셀러들에 비해 높은 평점은 아니다.

```python
# 베스트셀러의 평점은 어떠할까? 모든 책이 다 높을까? 낮은 평점임에도 베스트셀러인 책이 있을까?

print("distinct_name:",len(df['Name'].unique().tolist()), '권')

# drop_duplicates와 subset에 대한 기록 남기기

drop_name = df.drop_duplicates(subset=['Name'])
drop_name

sns.countplot(x=drop_name['User Rating'], palette='Set1').set_title('the user rating distribution of all bestselling books')
plt.show()
```

1. 총 351개의 작품들의 평점의 분포를 살펴보면 5권의 책을 제외한 346개의 책은 평점이 4.0이상으로 대부분은 높은 평점을 갖고 있음
2. 4.8 평점이 가장 많은 비중을 차지, 이어서 4.6, 4.7 순서대로 이어짐. 
2. 5권의 책들 중 3.3 / 3.6의 가장 낮은 두 권의 책은 무엇인지 살펴보자

```python
double_three = df[df.iloc[:, 3] == 3.3]
double_three

three_six = df[df.iloc[:, 3] == 3.6]
three_six
```

모두 아주 약한 음의 상관관계를 보인다. 리뷰가 많다고 해서 평점이 좋은 것이 아니며, 가격이 저렴하다 해서 평점이나 리뷰가 많은 것도 아니다.
