## 데이터 컬럼 설명

 1.  monthCode: 날짜 (월 단위)
 2.  serviceId: 서비스
 3.  userId: 유저
 4.  timeSpent: 방문 시간 (분 단위)
 5.  numberOfDays: 방문 일수 (일 단위)

## 데이터 탐색 및 정제 

1. 해당 데이터셋 읽어오고 컬럼 타입 변경
2. 데이터 테이블(DataFrame) 크기 확인
3. 결측치(null) 확인 및 값 대체
4. 'timePerDay'(유저의 일일 평균이용시간) 컬럼 추가(timePerDay = timeSpent / numberOfDays
5. index 중복값 확인
6. monthCode, serviceId를 기준으로 올림차순

```python
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# ================== 1 ======================
# monthCode, serviceId 컬럼은 category, userId는 string으로 타입 변경

df = pd.read_csv('serviceUsageMonthly.csv', sep = ',', encoding='utf-8')
df["monthCode"] = df["monthCode"].astype('category')
df["serviceId"] = df["serviceId"].astype('category')
df["userId"] = df["userId"].astype('str')
df.dtypes
```

```python
# column별 카테고리 확인

print("serviceId 카테고리 값", df.serviceId.values)
print("serviceId 카테고리 값", df.monthCode.values)
```

```python
# ================== 2 ======================
# df size 확인

df.shape
```

```python
# ================== 3 ======================
# 결측치 확인 

df.isnull().sum()

# timeSpent 결측치(2609개) 어떻게 처리할 것인가? => 0으로 처리 
```

```python
# 어떤 서비스에서 가장 많은 결측치가 나왔는지 확인

null_table = df[(df['timeSpent'].isnull())]
null_service = null_table.groupby('serviceId').count()
null_service = null_service.sort_values(by='monthCode', ascending=False)
null_service.head()
#'serviceId 1'에서 가장 timeSpent column에서 결측치가 많이 나옴. 
```

```python
# 0으로 결측치 대체

df['timeSpent'] = df['timeSpent'].fillna(0)

# 결측치 확인 ==> 모든 컬럼에서 0
df.isnull().sum()
```

```python
# ================== 4 ======================
# timePerDay 컬럼 추가

df['timePerDay'] = round((df.timeSpent / df.numberOfDays), 1)
df.shape     # (708844, 6)

df.head()
```

```python
# ================== 5 ======================
# 인덱스 중복값 확인

df.index.is_unique  
```

```python
# ================== 6 ======================
# monthCode, serviceId를 기준으로 올림차순

df = df.sort_values(by = ['monthCode', 'serviceId'])
df.head(10)
```

# A. 2020년의 전체 서비스 현황 요약

1. 2020년 '월별 중복 포함 유저 수'와 '월별 중복 제외 유저 수' 현황 비교
2. 2020년 서비스별 이용자 수 현황 및 이용자 수가 높은 서비스 파악
3. 2020년 서비스별 '일일 평균 이용시간' 현황(mean, median)
4. 2020년 서비스별 '한달 이용 일수' 현황(mean, median)
5. 2020년 유저별 이용 현황

```python
# ================== 1 ======================

# 월별 중복 포함 유저 수 현황

month_total_users = df.groupby('monthCode')[['userId']].count()
month_total_users = month_total_users.rename(columns = {'userId':'count_total_users'})
month_total_users


# 월별 중복 제외 유저 수 현황

month_distinct_users = df.groupby('monthCode')[['userId']].nunique()
month_distinct_users = month_distinct_users.rename(columns = {'userId':'count_distinct_users'})
month_distinct_users


# '월별 중복 포함 유저 수'와 '월별 중복 제외 유저 수' 현황 비교 및 두 칼럼의 차(difference_values) 값 비교

month_users = pd.concat([month_total_users, month_distinct_users], axis=1)
month_users['difference_values'] = month_users['count_total_users'] - month_users['count_distinct_users']
month_users
```

#### 1번 분석

- '월별 유저 수(count_total_users)'는 월마다 차이가 큰 편이나(최저값: 45,140명, 최고값: 73,184명), '월별 전체 유저 수(count_distinct_users)'는 '202004'를 제외하고 3만명 대에 그쳐 있다.
- 이는 월마다 새로운 유저의 유입이 많다기보다 기존 유저들이 특정 달에는 더 많은 서비스를 중복해서 이용했음을 추측해볼 수 있다. 그 특정 달은 (12, 8, 7, 1)월달로 추려지는데, 계절로 보면 여름과 겨울에 유난히 치우쳐져 있다. 계절별로 유저의 서비스 이용 횟수의 차이가 있다는 것을 알 수 있다.
- 4월달이 '월별 중복 포함 유저 수'와 '월별 중복 제외 유저 수' 모두 다 가장 값이 작다. 이는 무엇을 의미할까? 일년 중 가장 적은 유저수가 서비스를 이용했으며, 서비스 이용 횟수도 상대적으로 적었다는 것을 추측해볼 수 있다.
- 'difference_values' 에서는 12월달이 35,676으로 가장 높은 값이 나왔다. 이는 12월달의 유저들이 여러 개의 서비스를 이용했다고 추측해볼 수 있다. 그리고 (12, 8, 7, 1)월달이 모두 3만명이 넘는 것으로 보아 해당 달에는 유저들이 다양한 서비스를 중복 이용했음을 재차 추측해볼 수 있다. 

```python
# ================== 2 ======================

# 서비스별 중복 포함 유저 수 현황

service_total_users = df.groupby('serviceId')[['userId']].count()
service_total_users = service_total_users.rename(columns = {'userId':'count_total_users'})
service_total_users


# 서비스별 중복 제외 유저 수 현황

service_distinct_users = df.groupby('serviceId')[['userId']].nunique()
service_distinct_users = service_distinct_users.rename(columns = {'userId':'count_distinct_users'})
service_distinct_users


# '서비스별 중복 포함 유저 수'와 '서비스별 중복 제외 유저 수' 현황 비교

service_users = pd.concat([service_total_users, service_distinct_users], axis=1)
service_users
```

#### 2번 분석
- 전반적으로 서비스별 유저 수의 차가 심하다. 유저 수가 많은 상위 3개의 서비스는 두 컬럼 모두에서 (1, 19, 22) 서비스이며, 하위 3개의 서비스는 두 컬럼 모두에서 (18, 9) 서비스이고, 각각의 컬럼에서 8, 7 서비스인 것으로 나타났다.
- 대체로 'count_total_users'와 'count_distinct_users'은 비례관계인 것으로 보인다.
- 두 컬럼 값의 차이가 크면 클수록 해당 서비스를 이용한 유저는 여러 달에 걸쳐 해당 서비스를 이용하였음을 추측해볼 수 있는데, 1번과 19을 비교해보면 'count_total_users'에서는 19 더 높지만 'count_distinct_users'에서는 1번이 더 높다. 이는 해당 서비스를 이용하는 유저 자체는 1번 서비스가 많지만 여러 달에 걸쳐 서비스를 이용하는(지속성) 측면에서는 19번 서비스가 더 높다고 볼 수 있다.
- 값의 차이가 작으면 해당 서비스의 유저는 서비스 이용의 지속이 낮다고 추측해볼 수 있다.

```python
# ================== 3 ======================

# 서비스별 '일일 평균이용시간'의 평균값 현황

service_time_mean = round(df.groupby('serviceId')[['timePerDay']].mean(), 1)
service_time_mean = service_time_mean.rename(columns = {'timePerDay':'time_mean(m)'})


# 서비스별 '일일 평균이용시간'의 중앙값 현황

service_time_median = round(df.groupby('serviceId')[['timePerDay']].median(), 1)
service_time_median = service_time_median.rename(columns = {'timePerDay':'time_median(m)'})


# '서비스별 일일 평균이용시간의 평균값'과 '서비스별 일일 평균이용시간의 중앙값' 현황 비교

service_time = pd.concat([service_time_mean, service_time_median], axis=1)
service_time['difference_time(m)'] = service_time_mean['time_mean(m)'] - service_time_median['time_median(m)']
service_time
```

#### 3번 분석
- '일일 평균이용시간(timePerDay)' 컬럼을 이용해 서비스별 일일 평균이용시간의 평균값과 중앙값 현황을 분석.
- 평균값과 중앙값 두 개의 척도를 이용해 구하였고, 두 값의 차인 'difference_time(m)' 컬럼도 추가.
- 평균값으로 보면 '9' 서비스가 가장 일일 평균이용시간이 많고, '1' 서비스가 가장 일일 평균이용시간이 적다는 걸 알 수 있다. 
- 중앙값으로 보면 '20' 서비스가 가장 높고, '8' 서비스가 가장 낮다는 걸 알 수 있다.
- 그러나 '9' 서비스는 평균값과 중앙값의 차가 170.9로 가장 높은데, 이를 고려하면 해당 서비스 유저의 일일 평균이용시간에 이상치 값이 있을 것으로 추측된다. 최대값이 해당 서비스의 평균과 매우 큰 차이가 있을 것으로 보인다.
- '3' 서비스 역시 값의 차가 115.3으로 높은 편에 속하고 '9' 서비스와 비슷한 유형일 것으로 추측된다.
- 값의 차가 가장 적은 서비스는 14 서비스로 해당 서비스의 유저들은 일일 평균이용시간이 고르게 분포해 있을 것으로 예상해볼 수 있다.

```python
# ================== 4 ======================

# 서비스별 '한달 이용 일수'의 평균값 현황

service_day_mean = round(df.groupby('serviceId')[['numberOfDays']].mean(), 1)
service_day_mean = service_day_mean.rename(columns = {'numberOfDays':'days_mean(d)'})


# 서비스별 '한달 이용 일수'의 중앙값 현황

service_day_median = round(df.groupby('serviceId')[['numberOfDays']].median(), 1)
service_day_median = service_day_median.rename(columns = {'numberOfDays':'days_median(d)'})


# '서비스별 일일 평균이용시간의 평균값'과 '서비스별 일일 평균이용시간의 중앙값' 현황 비교

service_day = pd.concat([service_day_mean, service_day_median], axis=1)
service_day
```

#### 4번 분석
- days_median 컬럼 값이 1인 서비스는 '한달 동안 해당 서비스를 이용한 전체 유저'의 최소 50%는 하루만 해당 서비스를 이용한 것으로 해석할 수 있는데, 이때 '5', '8', '11', '18', '21'이 해당된다.
- days_median 컬럼에서 '22' 서비스가 값이 5로 가장 높은 값을 가지며, 이는 '한달 동안 해당 서비스를 이용한 전체 유저'의 50%는 5일 이상 해당 서비스를 이용한 것으로 해석할 수 있다.
- 또한, days_mean 컬럼에서도 '22' 서비스가 값이 7.9로 가장 높은 일수를 가지며, 이어서 '3' 서비스가 7.8로 높은 값이다.
- days_mean 컬럼에서 '18' 서비스가 값이 2.3으로 가장 낮은 일수로 days_median에서도 값이 1인데, 이는 해당 서비스의 유저들이 평균적으로 가장 낮은 일수를 방문하고 최소 50%는 한달에 1번 방문했던 것으로 볼 수 있다.

```python
# ================== 5 ======================

# 2020년 유저별 전체 이용 횟수 현황

users = df.groupby('userId')[['timeSpent']].count()
users = users.rename(columns = {'timeSpent':'count'})
users


users_sort = users.sort_values(by='count', ascending=False)
users_sort
print('size:', users_sort.shape)


# 유저들의 서비스 이용 횟수 분포 그래프

x = pd.Series(users_sort['count'], name='count_total_times')
sns.set(rc={'figure.figsize':(11,8)})
sns.distplot(x, hist=True, color='r')
```

4. 서비스별 이용 일수 랭킹 3위의 서비스 필터링 
  - 각 서비스의 한달 이용 일수 분포를 파악해보자
    - 월별 서비스의 이용 일수 평균, 표준편차 구한 후, 표준편차 작은 서비스는 무엇인지 살펴보기

```python
# ================== 1 ======================
# 어떤 서비스가 상대적으로 고정적인 유저 수가 보장될까?


# 서비스별 유저 수 랭킹3위의 서비스 필터링

service_user_ranking = service_total_users.sort_values(by='count_total_users', ascending=False)
print(service_user_ranking[:3])
print('=====================================')


# 월 & 서비스를 기준으로 그룹핑  ==> 해당 서비스의 월별 분포를 파악하기 위해

month_service_user = df.groupby(['monthCode', 'serviceId'])[['userId']].count()
df_m = month_service_user.unstack(level=0)


# 월별 서비스 유저 수 평균, 표준편차 구하고, 분포 및 변동계수 비교 

cv_list = []
for i in [1, 19, 22]:
    
    lst = df_m.loc[i]
    lst_stat = lst.describe()
    cv = round((lst_stat['std']/lst_stat['mean']), 2)
    cv_value = (i, "cv:", cv)
    
    cv_list.append(cv_value)
    print(i, "번 서비스\n", lst_stat)
    print('=====================================')
    
print("각 서비스별 변동계수 :", cv_list)
```

#### 1번 분석
- 서비스별 이용자수 상위 3개의 서비스는 '19', '1', '22'인 것으로 나타났다.
- 해당 서비스들의 요약통계를 살펴보면 '19' 서비스가 평균 역시 가장 높은 것으로 나타났고, 뒤를 이어 '1', '22' 서비스인 것으로 알 수 있다.
- 각 서비스들의 변동계수를 살펴보면 '19' 서비스가 0.1로 가장 낮은데 이는 매달 해당 서비스를 이용하는 유저의 수 분포의 격차가 상대적으로 작다는 것으로 파악할 수 있고, 반대로 '22' 서비스가 0.26으로 가장 높은데, 매달 해당 서비스 유저 수의 격차가 크다는 것으로 추측해볼 수 있다.

# C. 각 서비스 사이의 관계를 분석해 서비스 기획에 활용할 만한 게 있다면 제안하기

* 유저들 중 서비스 이용 횟수가 높은 유저들이 어떤 서비스를 이용했고, 해당 서비스 세트에 '연관규칙분석(Association Rules)'을 적용해 서비스간 연관성이 높은 세트가 무엇인지 확인하고 분석해 보고자 한다.(연관성이 높은 서비스세트를 생성해 추천 시스템으로 활용 가능)


### 연관규칙분석(Association Rules) 설명

#### "'A 서비스'를 이용하면 'B 서비스'도 이용하더라"를 알아보기 위해 Apriori 알고리즘을 이용해 분석 진행

- 모든 서비스들 간의 조합을 생성해 후보규칙을 모두 생성
- 후보규칙들 중 효과가 있는(자세히 살펴볼 가치가 있는) 규칙들을 뽑아내려면 3가지의 지표가 필요
  - Support(지지도): 규칙의 중요성에 대한 척도로 'A'와 'B' 서비스를 함께 이용한 경우의 수와 전체 경우의 수를 나눠 구함.
  - Confidence(신뢰도): 서비스들간의 연관성의 정도를 파악, 'A'를 이용할시 얼마나 'B'를 이용하는 것으로 이어지는지에 대한 것.
  - Lift(향상도): 서비스 A, B가 독립적인지, 상관이 있는지 파악하는 것으로 1보다 크면 항목간 양의 상관관계를 의미하고 1일 경우는 독립적, 0과 1 사이일 경우에는 항목간 음의 상관관계로 파악.
- 해당 개념을 이용해 서비스들 간에 효과가 있는 세트를 발견하면 추천시스템 혹은 마케팅으로 활용 가능.

```python
# 1) 유저들의 서비스 이용횟수를 카운팅 후, 36회 이상인 유저들만 필터링


users_sort = users.sort_values(by='count', ascending=False)
users_sort = users_sort[users_sort['count'] >= 36]
users_sort
```

#### 1) 설명

- 해당 조건으로 필터링하면 총 1841명이 추출. 1841명이 1년간 이용한 서비스 세트를 이용해 연관규칙분석을 진행
- 횟수를 36으로 정한 이유는 한달에 3가지 이상의 서비스를 이용(12 * 3)했을 때를 고려.

```python
# 2) 1841명의 유저 id만 추출

users_filtering = pd.DataFrame(users_sort.index)
users_filtering['userId']

range_users = users_filtering['userId']
range_users
```

```python
# 3) for loop을 이용해 모든 서비스 세트를 리스트에 담기

total_trans = []
count = 0
for i in range_users:
    find_user = df[df['userId'] == i]
    count_trans = list(find_user['serviceId'].unique())
    count += 1
    
#     print(count, count_trans)
    total_trans.append(count_trans)

# print(total_trans)
```

#### 3) 설명
- range_users를 for loop을 이용, 각 유저들이 이용한 서비스를 리스트로 담기(count_trans)
- total_trans(list)에 모든 count_trans 리스트를 담아서 apriori 알고리즘에 사용할 데이터셋으로 정제 
- 1841개의 리스트를 담은 하나의 리스트 total_trans

```python
# apriori 관련 패키지 'apyori' 설치 
 
import sys
!{sys.executable} -m pip install apyori
from apyori import apriori


# 4) 파라미터 설정(최소지지도 min_support, 최소신뢰도 min_confidence, 최소향상도 min_lift, 최소아이템개수 min_length, 최대아이템개수 max_length)

records = total_trans

associations = apriori(records, min_support=0.03, min_confidence=0.6, min_lift=2, min_length=2, max_length=3)
associations = list(associations)

print("모든 조건을 만족하는 서비스 세트는", len(associations), '개 입니다.')
# print(associations[0])
```

# apriori 패키지를 설치해야 해당 알고리즘이 작동합니다!

#### 4) 설명
- records는 total_trans(1841명의 유저들이 한 해 동안 이용한 서비스 세트) 입니다.
- 최소지지도는 0.03으로 잡았는데, 이는 ('하나의 서비스 세트'가 포함된 경우 / '모든 서비스 세트'가 포함된 경우)일 때, 0.03 이상이어야 한다는 것을 의미합니다. 
- 최소신뢰도는 0.6이상으로, 최소지지도를 만족하는 품목들을 대상으로 신뢰도를 구합니다. 서비스간 연관성('A' 서비스를 이용하면, 'B' 서비스도 이용하더라)이 중요하다고 생각해 0.6 이상으로 잡았습니다.
- 최소향상도는 1 이상이면 품목간 양의 상관관계를 의미하므로 2로 잡았습니다.
- 최소아이템개수는 2로 2개의 서비스 이상일 경우만 고려한다는 것이고, 최대아이템개수는 3으로 3개를 넘지 않는 서비스 세트를 고려했습니다. 4개 이상이 되면 파악하기가 복잡해지고 고려해야할 경우의 수가 많아져 3으로 잡았습니다.

```python
# 5) 조건에 만족하는 서비스 세트 상세하게 살펴보기

for item in associations:

    # first index of the inner list
    # Contains base item and add item
    pair = item[0] 
    print(pair)
    
    items = [x for x in pair]
    print("Rule: " + str(item[2][0][0]) + " -> " + str(item[2][0][1]))
    
    #second index of the inner list
    print("Support: " + str(round(item[1], 5)))

    #third index of the list located at 0th
    #of the third index of the inner list

    print("Confidence: " + str(round(item[2][0][2], 5)))
    print("Lift: " + str(round(item[2][0][3], 5)))
    print("=====================================")
```

#### 5) 분석

- '최소지지도 0.03, 최소신뢰도 0.6, 최소향상도 2, 2~3개 품목'을 만족하는 총 3개의 서비스 세트가 나왔습니다.
- 총 3개의 품목 중에서 향상도(lift)가 가장 높은 2번째 품목과 지지도(support)와 신뢰도(confidence)가 가장 높은 3번째 품목을 상세히 분석해보도록 하겠습니다.

- 서비스별 한달 이용일수는 두개의 feature(서비스별 유저 수와 서비스별 일일 평균이용시간) 모두와 약한 양의 상관관계로 나타났고, 이는 한달 이용일수와 유저 수(0.384), 한달 이용일수와 일일 평균이용시간(0.358) 모두 양의 비례 관계임을 알 수 있다.
- 이후부터는 두 개의 양의 상관관계를 이용해 각 유저 분류 및 분석을 진행하겠습니다.

```python
# ============== 2 ===============
# '서비스별 유저 수'와 '서비스별 이용 일수' 고려해 유저 분류 

df_count = df.groupby('numberOfDays')[['userId']].count()
df_count


# 서비스별 이용 일수를 그룹화(1-5)하기 위해 새로운 컬럼('group_days') 생성

days_df = df
days_df['group_days'] = 0


# 해당 컬럼 값을 '서비스별 이용 일수(numberOfDays)' 값을 분류해 그룹 아이디 부여

days_df.loc[days_df["numberOfDays"] == 1, "group_days"] = 1
days_df.loc[(days_df["numberOfDays"] <= 7) & (days_df["numberOfDays"] > 1), "group_days"] = 2
days_df.loc[(days_df["numberOfDays"] <= 14) & (days_df["numberOfDays"] > 7), "group_days"] = 3
days_df.loc[(days_df["numberOfDays"] <= 21) & (days_df["numberOfDays"] > 14), "group_days"] = 4
days_df.loc[(days_df["numberOfDays"] > 21), "group_days"] = 5


days_new_df = days_df.groupby('group_days')[['userId']].count()
days_new_df


df_users_days = days_df['userId'].groupby([days_df['serviceId'], days_df['group_days']]).count()
df_users_days = df_users_days.unstack(level=1).fillna(0)
df_users_days
```

#### 2) '서비스별 유저 수'와 '서비스별 이용 일수'  이용한 전략 및 제안


### 1번 그룹이 가장 많은 서비스 
: '1', '5', '6', '7', '8', '9', '10', '11', '13', '16', '18', '20', '21'
- 22개의 서비스 중 13개의 서비스가 유저의 한달 이용 일수가 하루인 경우가 총 유저 수 중 가장 많은 비율을 차지한다.
- 해당 서비스들은 유저의 유입이 있음에도 불구하고 서비스의 지속성이 매우 떨어진다는 것을 의미한다.
- 해당 서비스를 이용하는 유저들에게 서비스 이용의 어려움이나 불편함 점 등에 대한 설문조사를 진행해 어떤 요인들이 있는지 파악해 서비스를 개선해나갈 필요가 있다.

### 2번 그룹이 가장 많은 서비스 
: '2', '3', '4, '12', '14', '15', '17', '19', '22'
- 22개의 서비스 중 9개의 서비스가 유저의 한달 이용 일수가 2~7일로 일주일 이내인 경우에 해당된다.
- 해당 서비스들은 유저의 서비스의 지속성 1번 그룹보다는 더 낫지만, 이용 일수가 높지는 않다.
- 해당 서비스의 유저들에게는 서비스를 이용할 시, 이벤트나 깜짝 선물 등 서비스를 더 자주 이용하고자 하는 유인이 필요해 보인다.

### 5번 그룹의 절대 수가 높은 서비스
- '19', '22' 서비스는 10,000명 이상의 유저가 한달에 22일 이상을 이용하고 있다. 이는 서비스 이용 고정 유저가 다른 서비스에 비해 매우 많다. 해당 서비스의 고정 유저를 더 높이기 위한 전략이 필요해 보인다.
- '1', '3', '4', '12', '13', '14', '15' 서비스는 1,000명 이상의 유저가 한달에 22일 이상을 이용하고 있으며, 해당 서비스의 유저들 역시 고정 유저가 어느 정도 자리잡았다고 볼 수 있다. 하지만 앞서 본 서비스들과 마찬가지로 고정 유저를 확보하기 위한 전략이 필요하다.


```python
# ============== 3 ===============
# '서비스별 일일 평균이용시간'과 '서비스별 이용 일수' 고려해 유저 분류 

df_1 = df.groupby('timePerDay')[['userId']].count()


# 서비스별 이용 일수를 그룹화(1-5)하기 위해 새로운 컬럼('group_time') 생성

time_df = days_df
time_df['timePerDay'] = round(time_df['timeSpent'] / time_df['numberOfDays'], 2)
time_df['group_time'] = 0


# 해당 컬럼 값을 '서비스별 이용 일수(numberOfDays)' 값을 분류해 그룹 아이디 부여

time_df.loc[time_df["timePerDay"] <= 10, "group_time"] = 1
time_df.loc[(time_df["timePerDay"] <= 60) & (time_df["timePerDay"] > 10), "group_time"] = 2
time_df.loc[(time_df["timePerDay"] <= 100) & (time_df["timePerDay"] > 60), "group_time"] = 3
time_df.loc[(time_df["timePerDay"] <= 180) & (time_df["timePerDay"] > 100), "group_time"] = 4
time_df.loc[(time_df["timePerDay"] > 180), "group_time"] = 5


time_new_df = time_df.groupby('group_time')[['userId']].count()
time_new_df

df_days_time = time_df['userId'].groupby([time_df['group_days'], time_df['group_time']]).count()
df_days_time = df_days_time.unstack(level=0)
df_days_time
```

```python
# ============== 3 ===============
# '서비스별 일일 평균이용시간'과 '서비스별 이용 일수' 분류 결과를 이용한 전략 세우기

# 1) 일일 평균이용시간 100분 초과 & 한달 22일 이상 방문한 유저들에게 보상 전략
strategy_one = df_days_time[3:][5]
print("첫번째 전략 유저 수(중복):", strategy_one.sum())

# 2) 일일 평균이용시간 10분 초과 60분 이하 & 한달에 2~7일을 방문한 유저들에게 알람 혹은 선물 전략(이벤트성)
strategy_two = df_days_time[2][2]
print("두번째 전략 유저 수(중복):", strategy_two)
```

### 전략 2 (알람 혹은 선물 전략)
